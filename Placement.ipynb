```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("placement_data.csv")
```

```python
df
```

```python
df.sample(5)
```

```python
df.isnull().sum()
```

```python
plt.scatter(df["cgpa"],df["package"])
plt.xlabel("CGPA")
plt.ylabel("Package")
plt.show()
```

```python
x = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,Y_train)
```

```python
X_test.iloc[0]
```

```python
lr.predict(X_test.iloc[[0]])
```

```python
plt.scatter(df["cgpa"],df["package"],color="black")
plt.plot(X_train,lr.predict(X_train),color="red")
plt.xlabel("CGPA")
plt.ylabel("Package")
plt.show()
```

### Y = mx+B

```python
m = lr.coef_
b = lr.intercept_
```

```python
x = float(input("CGPA"))
y = m*x+b
print("Your package will be around :",y)
```
